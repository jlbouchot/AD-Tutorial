```python
import jax
import jax.numpy as jnp

# Define the quadratic function
@jax.jit
def quad_fun(x: float, a: float, b: float, c: float) -> float:
    """Calculates the value of a quadratic function a*x^2 + b*x + c."""
    return a * x * x + b * x + c


```

```python
# Main execution logic
# Some parameters
a = 2.0
b = -1.05
c = 0.5

x = 1.0
res = quad_fun(x, a, b, c)

print(f"Res = {res}")
```

### Calculate the gradient of `quad_fun` with respect to `x`

JAX's `grad` function can be used to automatically differentiate a function. By default, it differentiates with respect to the first argument. If you need to differentiate with respect to other arguments, you can specify `argnums`.

```python
# Calculate the gradient of quad_fun with respect to its first argument (x)
grad_quad_fun = jax.grad(quad_fun)

# Reuse the same parameters
a = 2.0
b = -1.05
c = 0.5
x = 1.0

# Compute the gradient at x = 1.0
grad_res = grad_quad_fun(x, a, b, c)

print(f"Gradient of quad_fun with respect to x at x={x} is: {grad_res}")
```
