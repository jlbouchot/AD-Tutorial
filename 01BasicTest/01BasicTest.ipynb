```python
import jax
import jax.numpy as jnp

# Define the quadratic function
@jax.jit
def quad_fun(x: float, a: float, b: float, c: float) -> float:
    """Calculates the value of a quadratic function a*x^2 + b*x + c."""
    return a * x * x + b * x + c


```

```python
# Main execution logic
# Some parameters
a = 2.0
b = -1.05
c = 0.5

x = 1.0
res = quad_fun(x, a, b, c)

print(f"Res = {res}")
```

### Calculate derivatives of `quad_fun` (forward mode using `jax.jvp()`)

1. [`jax.jvp`](https://jax.readthedocs.io/en/latest/jax.html#jax.jvp)
2. Computes $J \cdot v$ for a function $R^n \rightarrow R^m$.
3. `jax.jvp()` requires the input value for the primal function, which is the point where the derivatives are computed.
4. `jax.jvp()` requires a seed $v$. (For forward mode the shape of the seed must match the primal input.)
5. The code below obtains the entire Jacobian by calling `jax.jvp()` multiple times. It sets the input value to an arbitrary point and sets the seed to the Cartesian basis vectors.

`quad_fun(x, a, b, c)` has four scalar inputs, so we call `jax.jvp()` once per input argument:

- Initializing $\dot{x} = 1$ and $\dot{a} = \dot{b} = \dot{c} = 0$ yields $\dot{f} = \frac{\partial f}{\partial x}$.
- Initializing $\dot{a} = 1$ and $\dot{x} = \dot{b} = \dot{c} = 0$ yields $\dot{f} = \frac{\partial f}{\partial a}$.
- Initializing $\dot{b} = 1$ and $\dot{x} = \dot{a} = \dot{c} = 0$ yields $\dot{f} = \frac{\partial f}{\partial b}$.
- Initializing $\dot{c} = 1$ and $\dot{x} = \dot{a} = \dot{b} = 0$ yields $\dot{f} = \frac{\partial f}{\partial c}$.

```python
# Primal inputs
a = 2.0
b = -1.05
c = 0.5
x = 1.0
primals = (x, a, b, c)

# One jvp call per input argument (Cartesian basis tangents)
_, dfdx = jax.jvp(quad_fun, primals, (1.0, 0.0, 0.0, 0.0))
_, dfda = jax.jvp(quad_fun, primals, (0.0, 1.0, 0.0, 0.0))
_, dfdb = jax.jvp(quad_fun, primals, (0.0, 0.0, 1.0, 0.0))
_, dfdc = jax.jvp(quad_fun, primals, (0.0, 0.0, 0.0, 1.0))

val = quad_fun(x, a, b, c)

print(f"Value of quad_fun: {val}")
print(f"df/dx = {dfdx}")
print(f"df/da = {dfda}")
print(f"df/db = {dfdb}")
print(f"df/dc = {dfdc}")
print(f"Jacobian at (x, a, b, c) = ({x}, {a}, {b}, {c}): [{dfdx}, {dfda}, {dfdb}, {dfdc}]")
```
